```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_names = pd.read_csv("data/name.basics.tsv.gz", compression='gzip', delimiter='\t')
```

```python
df_crew = pd.read_csv("data/title.crew.tsv.gz", compression='gzip', delimiter='\t')
```

```python
df = pd.read_csv('FINAL_DATA/Film_Locations_Geocoded_FINAL.csv')
```

```python
#Extract only rows corresponding to focus data
unique_tconst = list(df.tconst.unique())
df_crew =  df_crew[df_crew['tconst'].isin(unique_tconst)]
```

```python
nconst_list = []
for index, row in df_crew.iterrows():
    directors = row['directors'].split(',')
    writers = row['writers'].split(',')
    nconst_list.extend(directors)
    nconst_list.extend(writers)
    
```

```python
nconst_df = pd.DataFrame(nconst_list, columns=['nconst'])
```

```python
nconst_list = list(nconst_df.nconst.unique())
```

```python
df_names = df_names[df_names['nconst'].isin(nconst_list)]
```

```python
for index, row in df_names.iterrows():
    professions = row['primaryProfession']
    professions = professions.replace(',',', ')
    professions = professions.replace('_',' ')
    professions = professions.title()
    df_names.at[index, 'primaryProfession'] = professions
```

```python
names_dict = {}
for index, row in df_names.iterrows():
    nconst = row['nconst']
    names_dict[nconst]=row['primaryName']
```

```python
#put each director in its own row and get names from names_dict
for index, row in df_crew.iterrows():
    directors = row['directors'].split(',')
    try:
        if len(directors) == 1:
            df_crew.at[index, 'director1_nconst'] = directors[0]
            df_crew.at[index, 'director1_name'] = names_dict[directors[0]]        
        elif len(directors) > 1:
            df_crew.at[index, 'director1_nconst'] = directors[0]
            df_crew.at[index, 'director2_nconst'] = directors[1]
            df_crew.at[index, 'director1_name'] = names_dict[directors[0]]  
            df_crew.at[index, 'director2_name'] = names_dict[directors[1]]  
    except:
        continue
    
```

```python
#put each writer in its own row and get names from names_dict
for index, row in df_crew.iterrows():
    writers = row['writers'].split(',')
    try:
        if len(writers) == 1:
            df_crew.at[index, 'writer1_nconst'] = writers[0]
            df_crew.at[index, 'writer1_name'] = names_dict[writers[0]]        
        elif len(writers) == 2:
            df_crew.at[index, 'writer1_nconst'] = writers[0]
            df_crew.at[index, 'writer2_nconst'] = writers[1]
            df_crew.at[index, 'writer1_name'] = names_dict[writers[0]]  
            df_crew.at[index, 'writer2_name'] = names_dict[writers[1]]  
        elif len(writers) > 2:
            df_crew.at[index, 'writer1_nconst'] = writers[0]
            df_crew.at[index, 'writer2_nconst'] = writers[1]
            df_crew.at[index, 'writer3_nconst'] = writers[2]
            df_crew.at[index, 'writer1_name'] = names_dict[writers[0]]  
            df_crew.at[index, 'writer2_name'] = names_dict[writers[1]]  
            df_crew.at[index, 'writer3_name'] = names_dict[writers[2]]  
    except:
        continue
```

```python
crew_dict = {}
for index, row in df_crew.iterrows():
    tconst = row['tconst']
    crew_dict[tconst] = {}
    crew_dict[tconst]['director1_nconst']=row['director1_nconst']
    crew_dict[tconst]['director1_name']=row['director1_name']
    crew_dict[tconst]['director2_nconst']=row['director2_nconst']
    crew_dict[tconst]['director2_name']=row['director2_name']
    crew_dict[tconst]['writer1_nconst']=row['writer1_nconst']
    crew_dict[tconst]['writer1_name']=row['writer1_name']
    crew_dict[tconst]['writer2_nconst']=row['writer2_nconst']
    crew_dict[tconst]['writer2_name']=row['writer2_name']
    crew_dict[tconst]['writer3_nconst']=row['writer3_nconst']
    crew_dict[tconst]['writer3_name']=row['writer3_name']
    
    
```

```python
for index, row in df.iterrows():
    tconst = row['tconst']
    try:
        df.at[index, 'director1_nconst'] = crew_dict[tconst]['director1_nconst']
        df.at[index, 'director1_name'] = crew_dict[tconst]['director1_name']
        df.at[index, 'director2_nconst'] = crew_dict[tconst]['director2_nconst']
        df.at[index, 'director2_name'] = crew_dict[tconst]['director2_name']
        df.at[index, 'writer1_nconst'] = crew_dict[tconst]['writer1_nconst']
        df.at[index, 'writer1_name'] = crew_dict[tconst]['writer1_name']
        df.at[index, 'writer2_nconst'] = crew_dict[tconst]['writer2_nconst']
        df.at[index, 'writer2_name'] = crew_dict[tconst]['writer2_name']
        df.at[index, 'writer3_nconst'] = crew_dict[tconst]['writer3_nconst']
        df.at[index, 'writer3_name'] = crew_dict[tconst]['writer3_name']
    except:
        continue
```

```python
#get unique titles
titles_list = []
for index, row in df_names.iterrows():
    titles = row['knownForTitles'].split(',')
    titles_list.extend(titles)

titles_df = pd.DataFrame(titles_list, columns=['title'])
titles_list = list(titles_df.title.unique())
titles_list
```

```python
df_titles = pd.read_csv("data/title.basics.tsv.gz", compression='gzip', delimiter='\t')
```

```python
df_titles = df_titles[df_titles['tconst'].isin(titles_list)]
```

```python
title_dict = {}
for index, row in df_titles.iterrows():
    tconst = row['tconst']
    title = row['primaryTitle']
    title_dict[tconst] = title
```

```python
title_dict
```

```python
for index, row in df_names.iterrows():
    title_names_list = []
    titles = row['knownForTitles'].split(',')
    try:
        for title in titles:
            name = title_dict[title]
            title_names_list.append(name)
        df_names.at[index, 'KnowForTitleNames'] = ', '.join(title_names_list)
    except:
        continue

    
```
