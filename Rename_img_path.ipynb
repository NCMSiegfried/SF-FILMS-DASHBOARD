```python
import os
import pandas as pd
```

```python
df = pd.read_csv('FINAL_DATA/Film_Locations_Geocoded_FINAL.csv')
```

```python
def remove_year_and_poster(name):
    words = name.split() 
    return ' '.join(words[:-2]) 
```

```python
title_tconst_dict = {}
for index, row in df.iterrows():
    title = row['Title']
    tconst = row['tconst']
    title_tconst_dict[title] = tconst
```

```python
path = 'map/images'
for item in os.listdir(path):
    full_path = os.path.join(path, item)
    title = remove_year_and_poster(item)
    try:
        new = title_tconst_dict[title]
        new_path = os.path.join(path, new)
        os.rename(full_path, new_path)
    except:
        continue
```
