```python
import regex as re
for index, row in df.iterrows():
    title = row.Title
    if 'Season' in title:
        df.at[index, 'Season'] = re.search(r'Season\s+(\d+)', title, re.IGNORECASE).group(1)
        if 'ep' in title:
            df.at[index, 'Episode'] = re.search(r'ep\s*(\d+)', title, re.IGNORECASE).group(1)
        elif 'Pilot' in title:
            df.at[index, 'Episode'] = 1
        if 'Murder in the First' in title:
            df.at[index, 'Title'] = 'Murder in the First'
        elif 'Chance' in title:
            df.at[index, 'Title'] = 'Chance'
        elif 'Goliath' in title:
            df.at[index, 'Title'] = 'Goliath'
        elif 'Sense8' in title:
            df.at[index, 'Title'] = 'Sense8'
        elif 'Ballers' in title:
            df.at[index, 'Title'] = 'Ballers'
        elif 'Silicon Valley' in title:
            df.at[index, 'Title'] = 'Silicon Valley'
        elif 'Blindspotting' in title:
            df.at[index, 'Title'] = 'Blindspotting'
        elif 'Looking' in title:
            df.at[index, 'Title'] = 'Looking'
        else:
            print(title)
```
