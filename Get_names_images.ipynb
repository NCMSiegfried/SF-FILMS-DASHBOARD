```python
from bing_image_downloader import downloader
import pandas as pd
```

```python
df_names = pd.read_csv('FINAL_DATA/Names_FINAL.csv')
```

```python
df_names
```

```python
for index, row in df_names.iterrows():
    query = f"{row.primaryName} IMDB"
    downloader.download(query, limit=1,  output_dir='map/images', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)

```

```python
names_dict = {}
for index, row in df_names.iterrows():
    ncost = row.nconst
    name = row.primaryName
    names_dict[name] = ncost
```

```python
import os
```

```python
directory = 'map/images'

for filename in os.listdir(directory):
    old_file_path = os.path.join(directory, filename)
    if filename[-5:] == ' IMDB':
        nconst = names_dict[filename[:-5]]
#     if os.path.isfile(file_path):
        print(f"File: {file_path}")
        new_file_path = os.path.join(directory, nconst)
        os.rename(old_file_path, new_file_path)
```
