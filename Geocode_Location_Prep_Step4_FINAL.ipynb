```python
import pandas as pd
import os, math, re
```

```python
df = pd.read_csv("data/Film_Locations_Geocoded_Step5.csv")
pd.set_option('display.max_columns', 500)
```

```python
df['lat_final2'] = None
df['lon_final2'] = None
```

```python
df[df.verified2=='N']
```

```python
for index, row in df.iterrows():
    verified = row['verified2']
    coord1 = row['coordinates_final']
    lat1 = row['lat_final']
    lon1 = row['lon_final']
    if verified == "N":
        coord = row['coordinates_final2']
        df.at[index, 'lat_final2'] = coord.split(',')[0]
        df.at[index, 'lon_final2'] = coord.split(',')[1]
    elif verified == "Y":
        df.at[index, 'coordinates_final2'] = coord1
        df.at[index, 'lat_final2'] = lat1
        df.at[index, 'lon_final2'] = lon1
```
