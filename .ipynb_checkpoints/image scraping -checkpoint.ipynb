```python
from icrawler.builtin import GoogleImageCrawler, BaiduImageCrawler, BingImageCrawler
import pandas as pd
import os, math, re
import logging
```

```python
df = pd.read_csv("FINAL_DATA/Film_Locations_Geocoded_FINAL.csv")
```

```python
df['img_name'] = None
```

```python
movies = []
for index, row in df.iterrows():
    title = row['Title']
    year = row['Release Year']
    search = f"{row.Title} {year}"
    movies.append(search)
```

```python
from collections import Counter
from itertools import groupby
```

```python
movies_unique = [k for k, g in groupby(sorted(movies))]
```

```python
len(movies_unique)
```

```python
movies_unique
```

```python
from bing_image_downloader import downloader

for movie in movies_unique:
    query = f'{movie} poster'
    downloader.download(query, limit=1,  output_dir='map/images', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
    
```

```python
os.getcwd()
```

```python
image_dir = 'map/images'

for filename in os.listdir(image_dir):
    if filename != '.DS_Store':
#         print(filename)
        image_path = os.path.join(image_dir, filename)
        images = os.listdir(image_path)

        for image in images:
            if image !='.DS_Store':
                if image.split('.')[1]!='jpg':
                    #print(filename+ ': '+image.replace(image, 'Image_1.jpg'))
                    #print(f'{image_path}/{image}')
                    old = f'{image_path}/{image}'
                    new = f"{image_path}/{image.replace(image, 'Image_1.jpg')}"
                    os.rename(old, new)
```

```python
df2 = pd.read_csv("FINAL_DATA/Film_Locations_Geocoded_FINAL.csv")
```

```python
movies_unique= df2['Title'].unique()
movies_unique
```

```python
years_unique= df2['Release Year'].unique()
years_unique
```

```python

# Filter the dataframe to keep only rows where column_name is in desired_values
filtered_df = df[df['originalTitle'].isin(movies_unique)]
```

```python
filtered_df
```

```python
df2['title_year'] = df2['Title'] + str(df2['Release Year'])
```

```python
df2
```

```python
for index, row in df2.iterrows():
    title = row['Title']
    year = row['Release Year']
    df2.at[index, 'title_year'] = f'{title} {year}'
```

```python
df2
```

```python
filtered_df
```

```python
for index, row in filtered_df.iterrows():
    title = row['primaryTitle']
    year = row['startYear']
    filtered_df.at[index, 'title_year'] = f'{title} {year}'
```

```python
filtered_df
```

```python
filtered_df_dict = {}
```

```python
for index, row in filtered_df.iterrows():
    title_year = row['title_year']
    title_type = row['titleType']
    tconst = row['tconst']
    genres = row['genres']
    filtered_df_dict[title_year] = {}
    filtered_df_dict[title_year]['tconst'] = tconst
    filtered_df_dict[title_year]['type'] = title_type
    filtered_df_dict[title_year]['genres'] = genres
```

```python
filtered_df_dict
```

```python
df2['tconst'] = None
df2['type'] = None
df2['genres'] = None
```

```python
for index, row in df2.iterrows():
    title_year = row['title_year']
    if title_year in filtered_df_dict.keys():
        df2.at[index, 'tconst'] = filtered_df_dict[title_year]['tconst']
        df2.at[index, 'type'] = filtered_df_dict[title_year]['type']
        df2.at[index, 'genres'] = filtered_df_dict[title_year]['genres']
    else:
        continue
```
