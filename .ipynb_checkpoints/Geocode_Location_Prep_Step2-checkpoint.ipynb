```python
bounding_box = [-123,37,-122,38]

for index, row in df[df['query_location'].notna()].iterrows():
    query = row['query_location']
    response = geocoder.forward(query, bbox=bounding_box, limit=2)
    if response.status_code == 200:
        if len(response.geojson()['features']) == 1:
            place_name1 = response.json()['features'][0]['place_name']
            lon1 = response.json()['features'][0]['center'][0]
            lat1 = response.json()['features'][0]['center'][1]
            relevance1 = response.json()['features'][0]['relevance']
            df.at[index, 'lat1'] = lat1
            df.at[index, 'lon1'] = lon1
            df.at[index, 'address1'] = place_name1
            df.at[index, 'geocode_relevance1'] = relevance1
        elif len(response.geojson()['features']) == 2:
            place_name1 = response.json()['features'][0]['place_name']
            lon1 = response.json()['features'][0]['center'][0]
            lat1 = response.json()['features'][0]['center'][1]
            place_name2 = response.json()['features'][1]['place_name']
            lon2 = response.json()['features'][1]['center'][0]
            lat2 = response.json()['features'][1]['center'][1]
            relevance1 = response.json()['features'][0]['relevance']
            relevance2 = response.json()['features'][1]['relevance']
            df.at[index, 'lat1'] = lat1
            df.at[index, 'lon1'] = lon1
            df.at[index, 'address1'] = place_name1
            df.at[index, 'lat2'] = lat2
            df.at[index, 'lon2'] = lon2
            df.at[index, 'address2'] = place_name2
            df.at[index, 'geocode_relevance1'] = relevance1
            df.at[index, 'geocode_relevance2'] = relevance2
        else:
            continue
```
